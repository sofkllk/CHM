```python
#Побудова графіка функції 
import math
import numpy as np 
import matplotlib.pyplot as plt 
from scipy import optimize 
from scipy.misc import derivative 

# область значень для x та y 
x_min, x_max = -1, 4 
y_min, y_max = -2, 3 
step = 0.01 

# створюємо масиви значень x та y 
x, y = np.meshgrid(np.arange(x_min, x_max, step), np.arange(y_min, y_max, step)) 

# рівняння системи 
eq1 = np.sin(y) + 2*x - 2 
eq2 = np.cos(x-1) + y - 0.7 


fig, ax = plt.subplots(figsize=(10, 10)) # створюємо графік 
ax.contour(x, y, eq1, levels=[0], colors='brown') #графік першого рівняння 
ax.contour(x, y, eq2, levels=[0], colors='green') # графік другого рівняння 

# налаштування графіка 
ax.set_xlim([x_min, x_max]) 
ax.set_ylim([y_min, y_max]) 

#підписи осей
ax.set_xlabel('x') 
ax.set_ylabel('y') 

ax.set_title('Графік системи рівнянь') #назва графіка
plt.grid(True) 
plt.show() 

# Розв'язок системи рівнянь методом простої ітерації 

x0 = 1.1 
y0 = -0.3 
delta = 0.1 
def f1(x): 
    return 0.7 - math.cos(x-1) #задаємо другу функцію 
def f2 (y): 
    return (2 - math.sin(y))/2 #задаємо першу  функцію
def iter (x,y,e): 
    xn = x 
    yn = y 
    xn1 =  f1(x)
    yn1 =  f2(y)
    n = 1 
    while ((abs(xn1-xn)>=e) & (abs(yn1-yn) >=e)): 
        xn = xn1 
        yn = yn1 
        xn1 = f2(yn) 
        yn1 = f1(xn) 
        n += 1 
    print ('Метод простої ітерації:') 
    print (f'x={xn} \ny= {yn} \nКількість ітерацій = {n}') 
iter(x0,y0,0.0001) 
def f3(x): #Задаємо функцію для перевірки 
    return math.sin(x[1])  + 2*x[0] -2, math.cos(x[0] - 1) + x[1] - 0.7 
s = optimize.root(f3, [0.,0.], method = 'hybr') #Перевірка розв*язку  

print ('Перевірка',s.x) 
```
